```python
# Importar librerias
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import scipy
from scipy import stats
```

```python
# Cargar datos desde el archivo CSV usando np.genfromtxt
data = np.genfromtxt('idealista_madrid.csv', delimiter=',', skip_header=1, dtype=None, encoding='utf-8')
```

```python
print(data)
```

```python
# Extraer las columnas necesarias y realizar las transformaciones
prices = np.array([float(row[0]) for row in data])
baths = np.array([float(row[1]) for row in data])
rooms = np.array([float(row[2]) for row in data])
sqft = np.array([float(row[3]) for row in data])
addresses = np.array([row[4].replace(', Madrid', '') for row in data])

# Imprimir solo los primeros cinco registros
for i in range(5):
    print(f"Precio: {prices[i]}, Baños: {baths[i]}, Habitaciones: {rooms[i]}, Metros Cuadrados: {sqft[i]}, Dirección: {addresses[i]}")
    
```

```python
# Media y mediana de prices
mean_price = np.mean(prices)
median_price = np.median(prices)

print(f"Media del precio {mean_price}. Mediana del precio {median_price}")
```

```python
# Media y mediana de baths
mean_baths = np.mean(baths)
median_baths = np.median(baths)

print(f"Media del baths {mean_baths}. Mediana del baths {median_baths}")
```

```python
# Media y mediana de rooms
mean_rooms = np.mean(rooms)
median_rooms = np.median(rooms)

print(f"Media del rooms {mean_rooms}. Mediana del rooms {median_rooms}")
```

```python
# Media y mediana de sqft
mean_sqft = np.mean(sqft)
median_sqft = np.median(sqft)

print(f"Media del sqft {mean_sqft}. Mediana del sqft {median_sqft}")
```

```python
# Máximo y mínimo
max_price, min_price = np.max(prices), np.min(prices)
max_baths, min_baths = np.max(baths), np.min(baths)
max_rooms, min_rooms = np.max(rooms), np.min(rooms)
max_sqft, min_sqft = np.max(sqft), np.min(sqft)
```

```python
# Imprimir máximos y mínimos
print(f"Precio máximo: {max_price}, Precio mínimo: {min_price}")
print(f"Baños máximo: {max_baths}, Baños mínimo: {min_baths}")
print(f"Habitaciones máximo: {max_rooms}, Habitaciones mínimo: {min_rooms}")
print(f"Metros Cuadrados máximo: {max_sqft}, Metros Cuadrados mínimo: {min_sqft}")
```

```python
# Resumen estadístico. 
# Resultados de media, mediana, máximo y mínimo
print("--- Estadísticas Generales ---")
print(f"Precio - Media: {mean_price}, Mediana: {median_price}, Máximo: {max_price}, Mínimo: {min_price}")
print(f"Baños - Media: {mean_baths}, Mediana: {median_baths}, Máximo: {max_baths}, Mínimo: {min_baths}")
print(f"Habitaciones - Media: {mean_rooms}, Mediana: {median_rooms}, Máximo: {max_rooms}, Mínimo: {min_rooms}")
print(f"Metros cuadrados - Media: {mean_sqft}, Mediana: {median_sqft}, Máximo: {max_sqft}, Mínimo: {min_sqft}")
```

```python
# Cuartiles e IQR
q1_price = np.percentile(prices, 25)
q2_price = np.percentile(prices, 50)
q3_price = np.percentile(prices, 75)
iqr_price = q3_price - q1_price
```

```python
# Imprimir cuartiles
print(f"Precio - Primer Cuartil (Q1): {q1_price}")
print(f"Precio - Segundo Cuartil (Q2 - Mediana): {q2_price}")
print(f"Precio - Tercer Cuartil (Q3): {q3_price}")
print(f"Precio - Rango Intercuartílico (IQR): {iqr_price}")
```

```python
# Filtrar el 20% más caro y el 20% más barato
top_20 = np.percentile(prices, 80)
bottom_20 = np.percentile(prices, 20)

top_20_prices = prices[prices >= top_20]
bottom_20_prices = prices[prices <= bottom_20]

```

```python
# Imprimir filtro del 20%
print("Precios del 20% más caro:")
print(top_20_prices)

print("\nPrecios del 20% más barato:")
print(bottom_20_prices)
```

```python
# Filtrar barrios del 20% más barato
bottom_20_addresses = addresses[prices <= bottom_20]

print("Direcciones de las propiedades en el 20% más barato:")
print(bottom_20_addresses)
```

```python
# Moda address
def mode(nums):
    """Calcula la moda de una lista de números.

    Args:
        nums: Una lista de números.

    Returns:
        La moda de la lista.
    """

    counts = {}
    for num in nums:
        counts[num] = counts.get(num, 0) + 1

    max_count = max(counts.values())
    return [num for num, count in counts.items() if count == max_count]

    print(num)
    
mode(addresses)
```

```python
# Filtrar barrios del 20% más caros
top_20_addresses = addresses[prices <= top_20]

print(f"La moda de los 20 barrios más baratos: {mode(bottom_20_addresses)}")
print(f"La moda de los 20 barrios más caros  : {mode(top_20_addresses)}")
```

```python
# Dispersión, varianza y desviasión standar
variance_price = np.var(prices)
std_dev_price = np.std(prices)

print(f"La varianza: {variance_price} y la desviasión Standar: {std_dev_price}")
```

```python
# Filtro de outliers usando Tukey
lower_tukey = q1_price - 1.5 * iqr_price
upper_tukey = q3_price + 1.5 * iqr_price
outliers_tukey = (prices < lower_tukey) | (prices > upper_tukey)

print(f"los outliers son: {outliers_tukey}")
```

```python
# Visualización de outliers
plt.figure(figsize=(10, 6))
plt.scatter(range(len(prices)), prices, label='Datos', color='gray')
plt.axhline(lower_tukey, color='red', linestyle='--', label='Límite Inferior (Tukey)')
plt.axhline(upper_tukey, color='blue', linestyle='--', label='Límite Superior (Tukey)')
plt.title('Detección de Outliers con Tukey')
plt.xlabel('Índice')
plt.ylabel('Precio (en euros)')
plt.legend()
plt.show()
```

```python
# Matriz de Correlación
data_matrix = np.array([prices, baths, rooms, sqft]).T
correlation_matrix = np.corrcoef(data_matrix, rowvar=False)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', xticklabels=['Precio', 'Baños', 'Habitaciones', 'M2'], yticklabels=['Precio', 'Baños', 'Habitaciones', 'M2'], cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

```python
# Estandarizar

def estandarizar_datos(data):
  """Estandariza un conjunto de datos.

  Args:
    data: Un array NumPy con los datos a estandarizar.

  Returns:
    Un array NumPy con los datos estandarizados.
  """

  media = np.mean(data, axis=0)
  desviacion_estandar = np.std(data, axis=0)
  data_standardized = (data - media) / desviacion_estandar
  return data_standardized

data_standardized = estandarizar_datos(correlation_matrix)
print(data_standardized)
```

```python
# Visualización de datos estandarizados
plt.figure(figsize=(10, 6))
plt.boxplot(data_standardized, labels=['Precio', 'Baños', 'Habitaciones', 'M2'])
plt.title('Datos Estandarizados')
plt.ylabel('Valor Estandarizado')
plt.show()
```

```python
# Asimetría y curtosis
# Cálculo de la asimetría
asimetria_prices = np.mean(((prices - mean_price) / std_dev_price) ** 3)

# Cálculo de la curtosis
curtosis_prices = np.mean(((prices - mean_price) / std_dev_price) ** 4) - 3 

# Imprimir los resultados
print(f"Asimetría (Precios): {asimetria_prices}")
print(f"Curtosis (Precios): {curtosis_prices}")
```

```python
# Contraste de hipótesis
# Hipótesis 1: Casas de un barrio X vs. Y
barrio_x = prices[addresses == 'Recoletos']
barrio_y = prices[addresses == 'Hortaleza']
stat1, p_value1 = stats.ttest_ind(barrio_x, barrio_y, equal_var=False)
print(f"Contraste de Hipótesis 1: Recoletos vs. Hortaleza")
print(f"Estadístico t: {stat1}, p-value: {p_value1}")
```

```python
# Hipótesis 2: Casas con >= 3 baños vs. 1-2 baños
group_3_or_more = prices[baths >= 3]
group_1_or_2 = prices[(baths == 1) | (baths == 2)]
stat2, p_value2 = stats.ttest_ind(group_3_or_more, group_1_or_2, equal_var=False)
print(f"Contraste de Hipótesis 2: Casas con >= 3 baños vs. 1-2 baños")
print(f"Estadístico t: {stat2}, p-value: {p_value2}")
```
