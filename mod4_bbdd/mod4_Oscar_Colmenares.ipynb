### Funciones para conectarme a la bbdd y crear cada tabla

```python
# Función para conectar a la base de datos
def conectar_db():
    return con.connect(
        host="localhost",
        port="3306",
        user="root",
        password="admin",
        database="supermercado"
    )
```

```python
# Función genérica para ejecutar queries de creación de tablas
def ejecutar_query(query, mensaje):
    try:
        mydb = conectar_db()
        mycursor = mydb.cursor(buffered=True)  # Cursor bufferizado
        mycursor.execute(query)
        resultados = mycursor.fetchall()
        print(mensaje)
        for resultado in resultados:
            print(resultado)
    except Exception as e:
        print(f"Error: {e}")
    finally:
        if mycursor:
            mycursor.close()
        if mydb:
            mydb.close()
```

```python
# Función para crear la tabla 'tiendas'
def crear_tabla_tiendas():
    query = """
    CREATE TABLE IF NOT EXISTS tiendas (
        id_tienda INT AUTO_INCREMENT PRIMARY KEY,
        nombre_tienda VARCHAR(100),
        direccion VARCHAR(255),
        ciudad VARCHAR(100)
    );
    """
    ejecutar_query(query, "Tabla 'tiendas' creada correctamente.")
    
# Función para crear la tabla 'empleados'
def crear_tabla_empleados():
    query = """
    CREATE TABLE IF NOT EXISTS empleados (
        id_empleado INT AUTO_INCREMENT PRIMARY KEY,
        nombre_empleado VARCHAR(100),
        puesto ENUM('Cajero', 'Gerente', 'Reponedor', 'Vendedor'),
        id_tienda INT,
        FOREIGN KEY (id_tienda) REFERENCES tiendas(id_tienda)
    );
    """
    ejecutar_query(query, "Tabla 'empleados' creada correctamente.")


# Función para crear la tabla 'categorias'
def crear_tabla_categorias():
    query = """
    CREATE TABLE IF NOT EXISTS categorias (
        id_categoria INT AUTO_INCREMENT PRIMARY KEY,
        nombre_categoria VARCHAR(100)
    );
    """
    ejecutar_query(query, "Tabla 'categorias' creada correctamente.")


# Función para crear la tabla 'productos'
def crear_tabla_productos():
    query = """
    CREATE TABLE IF NOT EXISTS productos (
        id_producto INT AUTO_INCREMENT PRIMARY KEY,
        nombre_producto VARCHAR(100),
        precio DECIMAL(10, 2),
        stock INT,
        id_categoria INT,
        FOREIGN KEY (id_categoria) REFERENCES categorias(id_categoria)
    );
    """
    ejecutar_query(query, "Tabla 'productos' creada correctamente.")


# Función para crear la tabla 'clientes'
def crear_tabla_clientes():
    query = """
    CREATE TABLE IF NOT EXISTS clientes (
        id_cliente INT AUTO_INCREMENT PRIMARY KEY,
        first_name VARCHAR(100),
        last_name VARCHAR(100),
        email VARCHAR(255),
        codigo_postal VARCHAR(20)
    );
    """
    ejecutar_query(query, "Tabla 'clientes' creada correctamente.")

# Función para crear la tabla 'ordenes'
def crear_tabla_ordenes():
    query = """
    CREATE TABLE IF NOT EXISTS ordenes (
        id_orden INT AUTO_INCREMENT PRIMARY KEY,
        id_cliente INT,
        id_empleado INT,
        fecha_orden DATETIME,
        metodo_pago ENUM('Tarjeta', 'Efectivo'),
        FOREIGN KEY (id_cliente) REFERENCES clientes(id_cliente),
        FOREIGN KEY (id_empleado) REFERENCES empleados(id_empleado)
    );
    """
    ejecutar_query(query, "Tabla 'ordenes' creada correctamente.")


# Función para crear la tabla 'detalle_orden'
def crear_tabla_detalle_orden():
    query = """
    CREATE TABLE IF NOT EXISTS detalle_orden (
        id_detalle INT AUTO_INCREMENT PRIMARY KEY,
        id_orden INT NOT NULL,
        id_producto INT NOT NULL,
        cantidad INT,
        precio_unitario DECIMAL(10, 2),
        descuento DECIMAL(10, 2) DEFAULT NULL,
        FOREIGN KEY (id_orden) REFERENCES ordenes(id_orden),
        FOREIGN KEY (id_producto) REFERENCES productos(id_producto)
    );
    """
    ejecutar_query(query, "Tabla 'detalle_orden' creada correctamente.")

```

```python
# Llamar a las funciones de creación de tablas en secuencia
def crear_todas_las_tablas():
    crear_tabla_tiendas()
    crear_tabla_empleados()
    crear_tabla_categorias()
    crear_tabla_productos()
    crear_tabla_clientes()
    crear_tabla_ordenes()
    crear_tabla_detalle_orden()

```

```python
# Ejecutar la creación de todas las tablas
crear_todas_las_tablas()
```

### Parte 3: Consultas SQL

#### 1. Listado de órdenes con detalles de cliente y empleado
#### * Muestra el ID de la orden, la fecha, el nombre del cliente, el nombre del empleado ####   que atendió la compra y el método de pago.
#### * Utiliza un JOIN entre las tablas ordenes, clientes y empleados.


```python
# Listado de órdenes con detalles de cliente y empleado
query = """
SELECT o.id_orden, o.fecha_orden, c.first_name, c.last_name, e.nombre_empleado, o.metodo_pago
FROM ordenes o
INNER JOIN clientes c ON o.id_cliente = c.id_cliente
INNER JOIN empleados e ON o.id_empleado = e.id_empleado;
"""
ejecutar_query(query, "Listado de órdenes con detalles de cliente y empleado")
```

```python
# Productos con stock bajo
query = """
SELECT p.nombre_producto, c.nombre_categoria, p.stock
FROM productos p
INNER JOIN categorias c ON p.id_categoria = c.id_categoria
WHERE p.stock < 100;
"""
ejecutar_query(query, "Productos con stock bajo")
```

```python
# Ventas totales por categoría
query = """
SELECT c.nombre_categoria, SUM(do.cantidad * do.precio_unitario) AS total_ventas
FROM detalle_orden do
INNER JOIN productos p ON do.id_producto = p.id_producto
INNER JOIN categorias c ON p.id_categoria = c.id_categoria
GROUP BY c.nombre_categoria;
"""
ejecutar_query(query, "Ventas totales por categoría")
    
```

```python
# Clientes con mayores gastos acumulados
query = """
SELECT 
    c.first_name, 
    c.last_name,
    SUM((do.cantidad * do.precio_unitario) - COALESCE(do.descuento, 0)) AS total_gastado
FROM
    clientes c
INNER JOIN ordenes o ON c.id_cliente = o.id_cliente
INNER JOIN detalle_orden do ON o.id_orden = do.id_orden
GROUP BY c.id_cliente
ORDER BY total_gastado DESC;
"""
ejecutar_query(query, "Ventas totales por categoría")
```

```python
# Empleados y número de órdenes gestionadas
query = """
SELECT 
    e.nombre_empleado, 
    e.puesto, 
    COUNT(o.id_orden) AS total_ordenes
FROM
    empleados e
INNER JOIN ordenes o ON e.id_empleado = o.id_empleado
GROUP BY e.id_empleado;
"""
ejecutar_query(query, "Ventas totales por categoría")
```

```python
# Ordenes filtradas por fecha y tienda
query = """
SELECT 
    o.id_orden, o.fecha_orden, t.nombre_tienda, c.first_name, c.last_name
FROM
    ordenes o
INNER JOIN empleados e ON o.id_empleado = e.id_empleado
INNER JOIN tiendas t ON e.id_tienda = t.id_tienda
INNER JOIN clientes c ON o.id_cliente = c.id_cliente
WHERE 
    o.fecha_orden BETWEEN '2024-01-01' AND '2024-01-31'
    AND t.nombre_tienda in (SELECT nombre_tienda FROM tiendas);
"""
ejecutar_query(query, "Ventas totales por categoría")

```

```python
# Ranking de productos más vendidos en cada tienda
query = """
SELECT 
    t.nombre_tienda, 
    p.nombre_producto, 
    SUM(do.cantidad) AS total_vendido
FROM 
    tiendas t
INNER JOIN empleados e ON t.id_tienda = e.id_tienda
INNER JOIN ordenes o ON e.id_empleado = o.id_empleado
INNER JOIN detalle_orden do ON o.id_orden = do.id_orden
INNER JOIN productos p ON do.id_producto = p.id_producto
GROUP BY t.id_tienda, p.id_producto
ORDER BY t.nombre_tienda, total_vendido DESC
LIMIT 3;

"""
ejecutar_query(query, "Ventas totales por categoría")
```

```python
# Valor total de las ventas por año
query = """
SELECT YEAR(o.fecha_orden) AS año, SUM(do.cantidad * do.precio_unitario) AS total_ventas
FROM ordenes o
INNER JOIN detalle_orden do ON o.id_orden = do.id_orden
GROUP BY YEAR(o.fecha_orden);
"""
ejecutar_query(query, "Ventas totales por categoría")
```

```python
# Clientes que han realizado al menos 3 pedidos
query = """
SELECT c.first_name, c.last_name
FROM clientes c
WHERE 3 <= (SELECT COUNT(*) FROM ordenes o WHERE c.id_cliente = o.id_cliente);
"""
ejecutar_query(query, "Ventas totales por categoría")
```
